```python
df_cordis = pd.read_csv(pathlib.Path("/export/usuarios_ml4ds/lbartolome/Repos/my_repos/UserInLoopHTM/notebooks/val_results_cordis.csv"))
```

```python
# Calculate the mean and standard deviation of cohr and disp_perc per ntopics
df_cordis = df_cordis.groupby('ntopics').agg({'cohr': ['mean', 'std'], 'disp_perc': ['mean', 'std']}).reset_index()

# Flatten the column names for easier access
df_cordis.columns = ['ntopics', 'cohr_mean', 'cohr_std', 'disp_perc_mean', 'disp_perc_std']

df_cordis
```

```python
df_cancer = pd.read_csv(pathlib.Path("/export/usuarios_ml4ds/lbartolome/Repos/my_repos/UserInLoopHTM/notebooks/val_results_cancer.csv"))
```

```python
# Calculate the mean and standard deviation of cohr and disp_perc per ntopics
df_cancer = df_cancer.groupby('ntopics').agg({'cohr': ['mean', 'std'], 'disp_perc': ['mean', 'std']}).reset_index()

# Flatten the column names for easier access
df_cancer.columns = ['ntopics', 'cohr_mean', 'cohr_std', 'disp_perc_mean', 'disp_perc_std']

df_cancer
```

```python
df_ai = pd.read_csv(pathlib.Path("/export/usuarios_ml4ds/lbartolome/Repos/my_repos/UserInLoopHTM/notebooks/val_results_ai.csv"))
```

```python
# Calculate the mean and standard deviation of cohr and disp_perc per ntopics
df_ai = df_ai.groupby('ntopics').agg({'cohr': ['mean', 'std'], 'disp_perc': ['mean', 'std']}).reset_index()

# Flatten the column names for easier access
df_ai.columns = ['ntopics', 'cohr_mean', 'cohr_std', 'disp_perc_mean', 'disp_perc_std']

df_ai
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize = (15, 4), sharey=True)
dfs = [df_cordis, df_cancer, df_ai]
titles = ["CORDIS", "S2-Cancer", "S2-AI"]
y_repr = ['cohrs_cv', 'cohrs_npmi', 'entropies']
y_labels = ['Coherence CV', 'Coherence NPMI', 'Entropy']
colors = ["#2D6187", "#28ABB9"]

labels_all = []
handles_all = []
ax2_shared = None

for i, (df, ax, title) in enumerate(zip(dfs, axs.flat, titles)):
    ax.errorbar(
        df['ntopics'],
        df['cohr_mean'],
        yerr=df['cohr_std'],
        fmt='o-',
        ecolor='black',
        capsize=5,
        color=colors[0],
        label='$C_{NPMI}$')
    
    ax2 = ax.twinx()
    ax2.errorbar(
        df['ntopics'],
        df['disp_perc_mean'],
        yerr=df['disp_perc_std'], 
        fmt='s-',
        ecolor='black',
        capsize=5,
        color=colors[1],
        label='Disp')

    # Set y-axis label only for the most-left plot
    if i % 3 == 0:
        ax.set_ylabel('$C_{NPMI}$')
    
    # Set y-axis label only for the most-rigth plot
    if i // 3 == 0 and (i + 1) % 3 == 0:
        ax2.set_ylabel('$DP$')
    
    ax.grid(True)
    ax.set_xlabel('K')
    ax.set_title(title)
    ax2.grid(True)
    
    ax.set_xlim(right=102)
    
    handles, labels = ax.get_legend_handles_labels()
    labels_all += labels
    handles_all += handles
    
    if title == "CORDIS" or title == "S2-Cancer":
        ax2.set_yticks([])
        

handles_all, labels_all = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

fig.legend(handles_all+handles2, labels_all+labels2, loc='upper center', ncol=2, bbox_to_anchor=(0.511, 1.25),
           frameon=True, shadow=False)

save_fig = "images/var_mallets.png"
fig.savefig(save_fig, dpi='figure', bbox_inches='tight')
```
