```python
import pandas as pd
import pathlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# Define the figure size and dpi
fig_width = 6.9  # inches
fig_height = 3.5  # inches
fig_dpi = 350

plt.rcParams.update({
    'figure.figsize': (fig_width, fig_height),
    'figure.dpi': fig_dpi,

    # Fonts
    'font.size': 18,

    # Axes
    'axes.labelsize': 12,
    'axes.titlesize': 12,
    'axes.linewidth': 1,
    'axes.grid': True,
    'grid.linestyle': ':',
    'grid.linewidth': 1,
    'grid.color': 'gray',

    # Legend
    'legend.fontsize': 8,
    'legend.frameon': True,
    'legend.framealpha': 0.8,
    'legend.fancybox': False,
    'legend.edgecolor': 'gray',
    'legend.facecolor': 'white',
    'legend.borderaxespad': 0.5,
    'legend.borderpad': 0.4,
    'legend.labelspacing': 0.5,

    # Lines
    'lines.linewidth': 2.0,
    'lines.markersize': 2,
    'axes.labelsize': 10,
    'axes.titlesize': 12,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
})
```

```python
df_cordis = pd.read_csv(pathlib.Path("/export/usuarios_ml4ds/lbartolome/Repos/my_repos/UserInLoopHTM/notebooks/val_results_cordis.csv"))
```

```python
# Calculate the mean and standard deviation of cohr and disp_perc per ntopics
df_cordis = df_cordis.groupby('ntopics').agg({'cohr': ['mean', 'std'], 'disp_perc': ['mean', 'std']}).reset_index()

# Flatten the column names for easier access
df_cordis.columns = ['ntopics', 'cohr_mean', 'cohr_std', 'disp_perc_mean', 'disp_perc_std']

df_cordis
```

```python
df_cancer = pd.read_csv(pathlib.Path("/export/usuarios_ml4ds/lbartolome/Repos/my_repos/UserInLoopHTM/notebooks/val_results_cancer.csv"))
```

```python
# Calculate the mean and standard deviation of cohr and disp_perc per ntopics
df_cancer = df_cancer.groupby('ntopics').agg({'cohr': ['mean', 'std'], 'disp_perc': ['mean', 'std']}).reset_index()

# Flatten the column names for easier access
df_cancer.columns = ['ntopics', 'cohr_mean', 'cohr_std', 'disp_perc_mean', 'disp_perc_std']

df_cancer
```

```python
df_ai = pd.read_csv(pathlib.Path("/export/usuarios_ml4ds/lbartolome/Repos/my_repos/UserInLoopHTM/notebooks/val_results_ai.csv"))
```

```python
# Calculate the mean and standard deviation of cohr and disp_perc per ntopics
df_ai = df_ai.groupby('ntopics').agg({'cohr': ['mean', 'std'], 'disp_perc': ['mean', 'std']}).reset_index()

# Flatten the column names for easier access
df_ai.columns = ['ntopics', 'cohr_mean', 'cohr_std', 'disp_perc_mean', 'disp_perc_std']

df_ai
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize = (11.5, 2.5), sharey=True)
dfs = [df_cordis, df_cancer, df_ai]
titles = ["CORDIS", "S2-Cancer", "S2-AI"]
y_repr = ['cohrs_cv', 'cohrs_npmi', 'entropies']
y_labels = ['Coherence CV', 'Coherence NPMI', 'Entropy']
colors = ["#2D6187", "#28ABB9", "#387838"]

labels_all = []
handles_all = []
ax2_shared = None

for i, (df, color, ax, title) in enumerate(zip(dfs, colors, axs.flat, titles)):
    ax.errorbar(
        df['ntopics'],
        df['cohr_mean'],
        yerr=df['cohr_std'],
        fmt='x-',
        ecolor='black',
        capsize=2,
        color='#36AE7C',
        label='$C_{NPMI}$')

    # Creating a twin axis on the right side for 'disp_perc'
    #ax2 = ax.twinx()
    #if ax2_shared is None:
    #    ax2_shared = ax2
    #else:
    #    ax2.get_shared_y_axes().join(ax2, ax2_shared)
    ax2 = ax.twinx()
    ax2_shared = ax2  # Set ax2_shared to ax2 for all cases
    ax2.errorbar(
        df['ntopics'],
        df['disp_perc_mean'],
        yerr=df['disp_perc_std'], 
        fmt='x-',
        ecolor='black',
        capsize=2,
        color='#187498',
        label='$DP$')
    #ax2.sharey(ax)
    
    # Set y-axis label only for the most-left plot
    if i % 3 == 0:
        ax.set_ylabel('$C_{NPMI}$')
    
    # Set y-axis label only for the most-rigth plot
    if i // 3 == 0 and (i + 1) % 3 == 0:
        ax2.set_ylabel('$DP$')
    
    ax.grid(True)
    ax.set_xlabel('K')
    ax.set_title(title)
    ax2.grid(True)
    
    ax.set_xlim(right=102)
    
    handles, labels = ax.get_legend_handles_labels()
    labels_all += labels
    handles_all += handles
    
    if title == "CORDIS" or title == "S2-Cancer":
        ax2.set_yticks([])
        

handles_all, labels_all = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

fig.legend(handles_all+handles2, labels_all+labels2, loc='upper center', ncol=2, bbox_to_anchor=(0.511, 1.1),
           frameon=True, shadow=False)

save_fig = "images/var_mallets.png"
fig.savefig(save_fig, dpi='figure', bbox_inches='tight')
```
